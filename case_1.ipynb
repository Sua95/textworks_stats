```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Real Estate Florida.csv",skiprows=4, header=0)
df = df.drop(columns = ["Unnamed: 0"])
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].astype(int)

df
```

```python
df_o = df.loc[(df["Type of property"] == "Office")]
df_h = df.loc[(df["Type of property"] == "House")]
df_a = df.loc[(df["Type of property"] == "Apartment")]
```

```python
plt.hist(df_o['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Office')
plt.show()
```

```python
plt.hist(df_h['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of House')
plt.show()
```

```python
plt.hist(df_a['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Appartment')
plt.show()

#Appartment[$297,669] Office [$116,937]
```

```python
df_o.plot.scatter(x='Area (ft.)', y='Price')
plt.xlabel('Area (ft.)')
plt.ylabel('Price')
plt.title('Price vs Area Office')
plt.show()
```

```python
df_h.plot.scatter(x='Area (ft.)', y='Price')
plt.xlabel('Area (ft.)')
plt.ylabel('Price')
plt.title('Price vs Area House')
plt.show()
```

```python
df_a.plot.scatter(x='Area (ft.)', y='Price')
plt.xlabel('Area (ft.)')
plt.ylabel('Price')
plt.title('Price vs Area Office')
plt.show()
```

```python
print(df_o['Price'].skew())
print(df_h['Price'].skew())
print(df_a['Price'].skew())
```

```python
print(df['Price'].mean())
print(df['Price'].median())
print(df['Price'].mode())
```

```python
print("Oficce")
print(df_o['Price'].cov(df['Area (ft.)']))
print(df_o['Price'].corr(df['Area (ft.)']))
```

```python
print("House")
print(df_h['Price'].cov(df['Area (ft.)']))
print(df_h['Price'].corr(df['Area (ft.)']))
```

```python
print("Appartment")
print(df_a['Price'].cov(df['Area (ft.)']))
print(df_a['Price'].corr(df['Area (ft.)']))
```
