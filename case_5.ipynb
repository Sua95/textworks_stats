```python
import pandas as pd
from scipy.stats import ttest_1samp


df = pd.read_csv('House Prices USA (1985-2020).csv', skiprows=3, header=0)
df = df.drop(columns = ["Unnamed: 0"])

df['House Price'] = df['House Price'].str.replace('$','')
df['House Price'] = df['House Price'].str.replace(',','')
df['House Price'] = df['House Price'].astype(float)

df
```

```python
df_pa = df.loc[(df["State"] == "PA")]
df_pa
```

```python
null_value = 350000
alpha = 0.05
t_stat, p_val = ttest_1samp(df_pa["House Price"], null_value)
if p_val < alpha:
    print("The p-value is", p_val, "which is less than the level of significance of", alpha)
    print("We reject the null hypothesis.")
else:
    print("The p-value is", p_val, "which is greater than the level of significance of", alpha)
    print("We fail to reject the null hypothesis.")
```
