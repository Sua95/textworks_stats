Count the change in your pocket. (Do not include bills.)
Randomly survey 30 classmates. Record the values of the change in Table

```python
import random as rn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dolars = []

for i in range(30):
    # generate a random number between 5 and 10
    random_number = round(rn.uniform(5, 10), 0)
    # add the random number to the data set
    dolars.append(random_number)
    
df = pd.DataFrame({"Dolars":dolars})

df
```

```python
# plot the histogram
plt.hist(df['Dolars'], bins=5, density=True, alpha=0.5)

# plot the KDE curve
sns.kdeplot(df['Dolars'], bw_method=0.20, color='red')

# add labels and title
plt.xlabel('Dolars')
plt.ylabel('Amount')
plt.title('Change in pocket')

# show the plot
plt.show()

print(df["Dolars"].mean())
print(df["Dolars"].std())
```

Repeat steps one through five of the section Collect the Data. with one exception. Instead of recording the change of 30 classmates, record the average change of 30 pairs.

Randomly survey 30 pairs of classmates.
Record the values of the average of their change in Table.

```python
pairs_dict = []

for i in range(30):
    random_pairs = df.sample(2)
    average_samples = random_pairs["Dolars"].mean()
    pairs_dict.append(average_samples)
    
pairs_df = pd.DataFrame({"Pairs Average": pairs_dict})

# plot the histogram
plt.hist(pairs_df['Pairs Average'], bins=5, density=True, alpha=0.5)

# plot the KDE curve
sns.kdeplot(pairs_df['Pairs Average'], bw_method=0.3, color='red')

# add labels and title
plt.xlabel('Dolars')
plt.ylabel('Amount')
plt.title('Change in pocket')

# show the plot
plt.show()


print(pairs_df['Pairs Average'].mean())
print(pairs_df['Pairs Average'].std())
```

```python
group_dict = []

for i in range(30):
    random_groups = df.sample(5)
    average_samples = random_groups["Dolars"].mean()
    group_dict.append(average_samples)
    
group_df = pd.DataFrame({"Group Average": group_dict})

# plot the histogram
plt.hist(group_df['Group Average'], bins=3, density=True, alpha=0.5)

# plot the KDE curve
sns.kdeplot(group_df['Group Average'], bw_method=0.5, color='red')

# add labels and title
plt.xlabel('Dolars')
plt.ylabel('Amount')
plt.title('Change in pocket')

# show the plot
plt.show()
```
